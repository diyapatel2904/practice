```python
import pandas as pd
import numpy as np

data=pd.read_csv("bmw_pricing_challenge.csv",usecols=['maker_key','model_key','fuel','paint_color','car_type','price'])
# print(data)
data.head()
```

```python
for col in data.columns:
    print(col,':', len(data[col].unique()),'labels')
```

```python
# .shape means making new columns will be created while performing one hot encoding
# as output shows it has created 94 columns in this
pd.get_dummies(data,drop_first=True).shape
```

```python
# data.fuel.value_counts().sort_values(ascending=False)
data.fuel.value_counts().sort_values(ascending=False).head(20)
```

```python
# list of most frequent catagories of the variable
top_10=[x for x in data.fuel.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# making that 10 binary varible

for label in top_10:
    data[label] = np.where(data['fuel']==label,1,0)
data[['fuel']+top_10].head(40)
```

```python
def one_hot_top_x(df, variable, top_x_labels):

    for label in top_x_labels:
        df[variable+'_'+label]=np.where(data[variable]==label,1,0)

data=pd.read_csv("bmw_pricing_challenge.csv",usecols=['maker_key','model_key','fuel','paint_color','car_type','price'])

one_hot_top_x(data,'fuel',top_10)
data.head()
```
