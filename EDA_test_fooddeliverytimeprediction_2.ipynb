```python
data = df['Delivery_Time_min']
```

```python
def detect_outliers(data):
    threshold =2   # Z-score threshold
    mean = np.mean(data)
    std = np.std(data)

    outliers = []
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)

    return outliers
```

```python
outliers = detect_outliers(data)
outliers
```
