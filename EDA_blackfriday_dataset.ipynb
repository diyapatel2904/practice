cleaning and prepaing the data for model training

A retail company "ABC Private limited" wants to understand the customer purchase behaviour (specialy,purchase amount)against various products of different categories.they have shared purchase summary of various customers for selected high volume products from last month.the datset also contains customer demographics(age,gender,marital status ,city_type,stay_in_current_city),product details (product_id and product category) and total purchase_amount from last month.

now , they want to build a model to predict the purchase amount of customer against various producta which will help them to create personalized offer for cutomers against different products.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df_train=pd.read_csv("blackFriday_train.csv")
df_train
```

```python
df_test=pd.read_csv("blackFriday_test.csv")
df_test
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
df.head()
```

```python
# handling the catagorical feature gender

df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
# handling the catagorical feature age

df['Age'].unique()
```

```python
# this is called as target ordinal encoding

df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```
