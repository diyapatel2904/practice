Polynomial Regression

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
X=6 * np.random.rand(200,1) - 3
y=0.8* X**2 + 0.9 * X + 2 + np.random.rand(200,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X,y,"b.")
plt.show()
```

```python
# applying polynomial lr 

poly = PolynomialFeatures(degree=2)

X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.transform(X_test)
print(X_train[0])
print(X_train_trans[0]) # [x^0 ,x^1,x^2]
```

```python
lr = LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred = lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
lr.coef_
```

```python
lr.intercept_
```

Logistics Regression

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
df.isnull().sum()
```

```python
df=df[df['species']!='setosa']
```

```python
df.head()
```

```python
df['species']=df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.head()
```

```python
# dividing dataset into independent and dependent feature

X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver='saga')
```

```python
from sklearn.model_selection import GridSearchCV

parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300],'l1_ratio': [0, 0.5, 1]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
# prediction

y_pred = classifier_regressor.predict(X_test)
```

```python
y_pred
```

```python
# accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
