```python
# feature selection-dropping constant features

import pandas as pd

data = pd.DataFrame({"A":[1,2,4,1,2,4],"B":[4,5,6,7,8,9],"C":[0,0,0,0,0,0],"D":[1,1,1,1,1,1],"E":[1,2,3,4,5,6]})
data.head()

# for larger dataset first do train,test,select then u have to apply on trained dataset
```

```python
# it will remove zero variance feature

from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(data)

VarianceThreshold(threshold=0)
var_thres.get_support() #output says that A,B are not zero variance columns and C,D are zero variance
```

```python
data.columns[var_thres.get_support()]
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for feature in constant_columns:
    print(feature)
```

```python
data.drop(constant_columns,axis=1)
```
